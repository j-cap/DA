```python
# generate some data
np.random.seed(42)
n_samples = 200
noise = 0.5* np.random.random(n_samples)

x = np.linspace(0,1,n_samples)
y = 10*np.exp(-(x-0.5)**2 / 0.01) + noise
x, y = x.reshape(-1,1), y.ravel()

x_crit = np.array([[0.5], [2.4]])

fig = px.scatter(x=x.ravel(), y=y.ravel())
fig.add_trace(go.Scatter(x=x_crit[0], y=x_crit[1], name="Fix Point")) 
```

```python
m1 = (("s(1)", "conc", 55, (0.1, 100), "quantile"),)
M1 = Model(description=m1)
```

```python
M1.fit(X=x, y=y)
```

```python
critical_point = [ [0.2, 2], ] #[0.6, -3], [0.8, -2] ]

m1 = (("s(1)", "peak", 50, (1, 1000), "quantile"),)

M1 = Model(description=m1)

M1.weighted_fit(X=x, y=y, critical_point=critical_point)
```

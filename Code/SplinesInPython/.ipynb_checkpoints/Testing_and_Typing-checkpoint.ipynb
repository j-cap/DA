# Extrapolation

```python
import numpy as np
import plotly.express as px
import plotly.graph_objects as go
from stareg import StarModel 
```

```python
# test with different dataset
np.random.seed(41)
x = np.random.random(250)
x.sort()
xmax, xmin = x.max(), x.min()
x = x
x = (x - xmin) / (xmax - xmin)
noise = np.random.random(250)*1.25
y = 2*np.exp(-(x - 0.4)**2 / 0.1) - 5.5*x**2 + noise
n_param = 25
descr = ( ("s(1)", "peak", n_param, (0.01, 1000), "equidistant"), )
M = StarModel(description=descr)
M.fit(X=x.reshape(-1,1), y=y)
```

```python
def normalized(x, xmin=xmin, xmax=xmax):
    return (x - xmin) / (xmax - xmin)
```

```python
# data normalization
x_ep = np.concatenate((np.linspace(-1,0,10), np.linspace(1, 2, 10)))
x_ep_norm = normalized(x_ep)
```

```python
def extrapolate(x_exp, type_="constant", model=None, depth=3):
    """
    Type is either "constant", "linear", "zero"
    
    depth: how many coefficients are taken into account for the linear extrapolation

    """
    direction = "left" if x_exp < 0 else "right"
    
    if direction == "left":
        y_boundary = model.predict_single_point(x_sp=0)
        k = np.mean(np.diff(model.coef_[:depth]))
    elif direction == "right":
        y_boundary = model.predict_single_point(x_sp=1)
        k = np.mean(np.diff(model.coef_[-depth:]))
       
    if type_ == "constant":
        y_extrapolate = y_boundary
    elif type_ == "linear":
        dx = x_exp-1 if x_exp > 0 else x_exp
        y_extrapolate = y_boundary + dx * k
    elif type_ == "zero":
        #  bspline to zero
        pass
    
    return y_extrapolate
```

```python
y_extra = []
for i in x_ep_norm:
    y_extra.append(extrapolate(x_exp=i, type_="linear", model=M, depth=8))
    
fig = go.Figure()
fig.add_trace(go.Scatter(x=x, y=y, name="data"))
fig.add_trace(go.Scatter(x=x, y=M.basis @ M.coef_, name="Fit"))
fig.add_trace(go.Scatter(x=x_ep,y =y_extra, mode="markers", name="extrapolation"))
```

```python
def bspline(x, knots, i, m=2):
    if m==-1:
        return (x >= knots[i]) & (x < knots[i+1]).astype(int)
    else:
        #print("m = ", m, "\t i = ", i)
        z0 = (x - knots[i]) / (knots[i+m+1] - knots[i])
        z1 = (knots[i+m+2] - x) / (knots[i+m+2] - knots[i+1])
        return z0*bspline(x, knots, i, m-1) + z1*bspline(x, knots, i+1, m-1)
    
```

```python
def right_exp_spp(knots, sp=0, coef_=None):

    idx = np.argwhere(knots >= sp)[0][0]
    s = []
    print("idx = ", idx)

    if idx >= len(knots)-4-3:
        # right boundary knots
        s.append(0.)
        print("right")
        [s.append(bspline(x=sp, knots=knots[idx-3:], i=i, m=2)) for i in range(3)]
        print(s)
    p = np.sum(np.array(s) * coef_[idx-4-3:idx-3])   
    return p
```
